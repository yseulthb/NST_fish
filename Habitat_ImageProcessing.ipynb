```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
import imageio
from PIL import Image

# # Set the directory path of caeruleum fish images
# path = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/caeruleum/JPEG/male/'

# # # Set the directory path of caeruleum habitat images (Trammel Creek)
# path = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/TrammelCreek/defWB/'
# path = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/TrammelCreek/div1-5_scaling/'
# path = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/e_gracile_BrushxUnknownCreeks/whitebalance/'
# path = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/e_gracile_BrushxUnknownCreeks/div1-5_scaling/'
path = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/e_olmstedi_PatuxentRiver_sand/whitebalance/'

files = os.listdir(path)
n_files = len(files)
print(n_files)

# HABITAT patch_dir = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/TrammelCreek/patches_3000x1000/'
# patch_dir = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/e_gracile_BrushxUnknownCreeks/patches_3000x1000/'
# patch_dir = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/e_gracile_BrushxUnknownCreeks/Habitat_patches_6500x2400/'
patch_dir = 'C:/Users/preinstalled/PycharmProjects/pythonProject/fish_images/e_olmstedi_PatuxentRiver_sand/Habitat_patches_6500x2400/'
```

```python
# Habitat images: scale down the white balanced images (size divided by n)
from PIL import Image
for im_ndx in (files): # (files[:-6]) only first image
    print('image name: {}'.format(im_ndx))
    image    = im_ndx
    img_path = os.path.join(path, image)
    image    = plt.imread(img_path)
    foo      = Image.open(img_path)
    print(foo.size)
    print('image dimensions: {}'.format(image.shape))
    new_width  = np.array(foo.size[0]/1.5)
    new_height = np.array(foo.size[1]/1.5)
    
    foo = foo.resize((new_width,new_height),Image.ANTIALIAS)
    print(foo.size)
    
    plt.figure()
    plt.imshow(foo)
    plt.show()
    
#     foo.save(patch_dir + 'tt_resized_div1_5_' + im_ndx[0:-4] + '.tif', optimize = True, quality=95)
```

```python
# Habitat images: take the scaled down images and crop them to have an image of the same size as fish images

for im_ndx in (files): # (files[:-6])only first image
    print('image name: {}'.format(im_ndx))
    image    = im_ndx
    img_path = os.path.join(path, image)
    image = Image.open(img_path)
    print(image.size)
    midY = int(image.size[1] / 2)
    print(midY)
    
    plt.figure()
    plt.imshow(image) 
    plt.axhline(y = midY, linestyle = '--', linewidth = 1, color = 'r')
    plt.show()     

    # Setting the points for cropped image 
    left = 1
    right = left + 3000
    top = int(1500) #-1500/2)
    print(top)
    bottom = top + 1000
    print(bottom)
  
    # Cropped image of above dimension 
    # (It will not change the original image) 
    im1 = image.crop((left, top, right, bottom))   
    
    # Shows the image in image viewer 
    plt.figure()
    plt.imshow(im1) 
    plt.show()
    
    print(im1.size)
    
    im1.save(patch_dir + 'cropped_3000x1000_middle_'+ im_ndx[0:-4] + '.jpg')    
```
